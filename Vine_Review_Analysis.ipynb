```python
# import dependencies 

import pandas as pd
```

```python
# read vine table

vine_table = pd.read_csv("vine_table.csv")
vine_table
```

```python
# Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is
# equal to or greater than 20 to pick reviews that are more likely to be helpful and to 
# avoid having division by zero errors later on.

total_vote_count_df = vine_table.loc[vine_table["total_votes"]>= 20]
total_vote_count_df
```

```python
# Filter the total_vote_count_df and create a new DataFrame to retrieve all the rows where the number of 
# helpful_votes divided by total_votes is equal to or greater than 50%

helpful_reviews_df = total_vote_count_df.loc[(total_vote_count_df["helpful_votes"]/total_vote_count_df["total_votes"])>=0.5]
helpful_reviews_df
```

```python
# Filter thehelpful_reviews_df that retrieves all the rows where a review was written as part of the Vine program (paid)
# Vine = Y

helpful_vine_reviews_df = helpful_reviews_df.loc[helpful_reviews_df["vine"]=="Y"]

helpful_vine_reviews_df

```

```python
# Filter thehelpful_reviews_df that retrieves all the rows where a review was written  NOT as part of the Vine program (unpaid)
# Vine = N

helpful_nonvine_reviews_df = helpful_reviews_df.loc[helpful_reviews_df["vine"]=="N"]

helpful_nonvine_reviews_df
```

```python
# Determine the total number of reviews, the number of 5-star reviews (vine)

vine_fivestar_df = helpful_vine_reviews_df.loc[helpful_vine_reviews_df["star_rating"]==5]

vine_fivestar_df
```

```python
len(vine_fivestar_df)
```

```python
nonvine_fivestar_df = helpful_nonvine_reviews_df.loc[helpful_nonvine_reviews_df["star_rating"]==5]

nonvine_fivestar_df
```

```python
len(nonvine_fivestar_df)
```

```python
# percentage of 5-star reviews for the paid

vine_fivestar_percent = ((len(vine_fivestar_df))/(len(helpful_reviews_df.loc[helpful_reviews_df["star_rating"]==5])))*100
vine_fivestar_percent
```

```python
# percentage of 5-star reviews for unpaid
nonvine_fivestar_percent = ((len(nonvine_fivestar_df))/(len(helpful_reviews_df.loc[helpful_reviews_df["star_rating"]==5])))*100
nonvine_fivestar_percent
```
